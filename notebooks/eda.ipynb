```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/WA_Fn-UseC_-Telco-Customer-Churn.csv')
print(df.shape)
df.head()
```

```python
sns.countplot(x='Churn', data=df, palette='Blues')
plt.title('Churn Distribution')
plt.show()
```

```python
df.dtypes
```

```python
cat_cols = df.select_dtypes(include='object').columns
for col in cat_cols:
    print(f"{col}: {df[col].unique()}")
```

```python
print(df['TotalCharges'].dtype)
print(df[df['TotalCharges'] == ' '])
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
print(f"Nulls after conversion: {df['TotalCharges'].isnull().sum()}")
```

```python
df.dropna(subset=['TotalCharges'], inplace=True)
print(df.shape)
```

"TotalCharges was imported as object dtype due to 11 rows containing whitespace instead of null. I used pd.to_numeric with errors='coerce' to expose them, then dropped those rows — losing less than 0.2% of data."

```python
sns.countplot(x='Contract', hue='Churn', data=df, palette='Blues')
plt.title('Churn by Contract Type')
plt.show()
```

```python
sns.histplot(data=df, x='tenure', hue='Churn', bins=30, palette='Blues')
plt.title('Churn by Tenure')
plt.show()
```

```python
sns.boxplot(x='Churn', y='MonthlyCharges', data=df, palette='Blues')
plt.title('Monthly Charges vs Churn')
plt.show()
```

```python
sns.heatmap(df[['tenure','MonthlyCharges','TotalCharges']].corr(), annot=True, cmap='Blues')
plt.title('Correlation Matrix')
plt.show()
```

"EDA revealed that contract type, tenure, and monthly charges are the strongest churn indicators. Month-to-month customers with high charges and low tenure are the highest risk segment."
